```python
from typing import Optional
class Employee(BaseModel) :
  id:int 
  name :str
  department : str 
  salary : Optional[float]=None #Optinal with default value 
  is_active : Optional[bool]=None #Optional with default True 
```

```python
emp1=Employee(id=1,name="John",department="IT")
print(emp1)
```

```python
emp1=Employee(id=1,name="John",department="IT",salary=60000,is_active=False) ## automatic typing happening it converting from int to float
print(emp1) 
```

```python
print(emp1.name)
print(emp1.id)
```

```python
from typing import List
class Classroom(BaseModel) :
  room_number : str
  students : List[str] #List of strings
  capacity : int 
```

```python
classroom=Classroom(room_number="a101",students=["alice","bob","charlie"],capacity=30)
```

```python
print(classroom.students)
```

```python
from pydantic import BaseModel 
class Address(BaseModel) :
  street : str 
  village : str 
  city : str 

class User(BaseModel) :
  name : str 
  age : int 
  address : Address
  
```

```python
user=User(
  name="rahuletyala",
  age=21,
  address={
    "street" : "masjid street",
    "village" : "jallapur",
    "city":"Hyderabad"
  }
)
```

```python
print(user.address.street)
```

```python
from pydantic import BaseModel,Field 
class Item(BaseModel) :
  name : str=Field(min_length=10,max_length=50)
  price : float=Field(gt=0,lt=1000) #greater than 0, less than or equal to 1000
  quantity : int=Field(gt=0,lt=120)
```

```python
item=Item(name="Bookdasjklakds",price=29.99,quantity=10)
```

```python
print(item)
```

```python
from pydantic import BaseModel,Field 

class User(BaseModel) :
  username : str=Field(...,description="unique username for the user")
  age : int=Field(default=18,description="User age, defaults to 18")
  email : str=Field(default_factory=lambda: "user@example.com",description="default email address")
```

```python
user1=User(username="alice")
print(user1)
```

```python
user2=User(username="bob",age=25,email="bob@domain.com")
print(user2)
```

```python
print(User.model_json_schema())
```
