```python
## load documents 
from langchain_community.document_loaders import TextLoader
```

```python
docs=TextLoader('speech.txt').load()
docs[0]
```

```python
##split into chunks
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=30)
```

```python
chunks=text_splitter.split_documents(docs)
chunks[0]
```

```python
## convert chunks into vectors
from langchain_openai import OpenAIEmbeddings
from pprint import pprint
embeddings=OpenAIEmbeddings(model="text-embedding-3-large",api_key=api_key,base_url="https://aicredits.in/v1")
```

```python
## store in vector database 
from langchain_community.vectorstores import Chroma
db=Chroma.from_documents(chunks,embeddings) 
db
```

```python
## retrive the result 
query="While normally used to facilitate communication with others"
retrieved_results=db.similarity_search(query)
pprint(retrieved_results)
```
